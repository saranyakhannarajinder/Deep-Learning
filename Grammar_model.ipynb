```python
import language_tool_python
from happytransformer import HappyTextToText, TTSettings

# Initialize the language tool for English
tool = language_tool_python.LanguageTool('en-US')

# Initialize the HappyTextToText model
happy_tt = HappyTextToText("T5", "t5-base")

# Define the settings for text generation
args = TTSettings(num_beams=5, min_length=1, max_length=200)

# Input text
input_text = ("in this scenarios when an mportant model is due the very next day "
              "and the resources are not as excpected the best we can do is just deploy "
              "the modeel first and then run tests on the the data real-time data is coming into "
              "the model and debug the model with those data and then it rained over different "
              "version and release the cleaner model after deploying the not so good model.")

# Check grammar and spelling
matches = tool.check(input_text)

# Extract mistakes and corrections
mistakes = []
corrections = []

for match in matches:
    mistakes.append(input_text[match.offset:match.offset + match.errorLength])
    corrections.append(match.replacements[0])

# Generate corrected text
corrected_text = tool.correct(input_text)

# Output the results
print("Original Text:\n", input_text)
print("\nMistakes Found:\n", mistakes)
print("\nCorrections:\n", corrections)
print("\nCorrected Text:\n", corrected_text)
```
